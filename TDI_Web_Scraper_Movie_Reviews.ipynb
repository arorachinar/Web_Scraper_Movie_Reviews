```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd

names=[]
years=[]
IMDB_Rating=[]
MetaScore_Rating=[]
Votes=[]
baseUrl="https://www.imdb.com/search/title?release_date=2010-01-01,2018-12-31&count=250"
```

```python

for page in range(1,50):
    if page==1:
        url=baseUrl
    else:
        url=baseUrl+"&page="+str(page)+"&ref_=adv_nxt"
    countPages=countPages+1    
    response=get(url)
    html_soup=BeautifulSoup(response.text,'html.parser') 
    movie_containers=html_soup.find_all('div',class_='lister-item mode-advanced')
    for container in movie_containers:
        if container.find('div',class_='ratings-metascore') is not None:
            name=container.h3.a.text 
            names.append(name)
            year=container.h3.find('span',class_=('lister-item-year text-muted unbold')).text
            years.append(year)
            imdb=container.strong.text
            IMDB_Rating.append(imdb)
            metascore=container.find('div',class_=('inline-block ratings-metascore')).span.text
            MetaScore_Rating.append(metascore)
            num_metric=container.find('span',attrs={'name':'nv'})
            votes=num_metric['data-value']
            Votes.append(votes)
        
```

```python


AllMovies=pd.DataFrame({'Movie_Name':names,'Release_Year':years,'IMDB_Rating':IMDB_Rating,'Metascore_Rating':metascore,'Votes_Received':Votes})
```
